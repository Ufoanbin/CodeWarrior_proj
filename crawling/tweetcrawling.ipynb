```python
stock_merge_df = []
for i in ceo_list:
    id_i = i['아이디']
    # print(id_i)
    ticker_i = i['주식기호']
    stock_df = pd.read_csv('..//dataSet//stock/'+ticker_i+'.csv')
    stock_df.insert(1, 'name', id_i)
    # print(stock_df)
    stock_merge_df.append(stock_df)
stock_merge_df = pd.concat(stock_merge_df)
stock_merge_df.rename(columns={'Adj Close': 'AdjClose'}, inplace=True)
stock_merge_df
```
